```python
import networkx as nx
import matplotlib.pyplot as plt

#import import_ipynb
#from processing import list_of_wordsInSent

import pickle
from matplotlib import pyplot as plt
```

```python
with open('_inputs/list_of_wordsInSent_TA.pkl', 'rb') as f:
    list_of_wordsInSent_TA = pickle.load(f)

with open('_inputs/list_of_wordsInSent_RA.pkl', 'rb') as f:
    list_of_wordsInSent_RA = pickle.load(f)
```

```python
list_of_wordsInSent_TA;
```

```python
def graph(sent):
    nodes = list(sent.keys())
    node_labels = list(sent.values())
    
    G = nx.DiGraph()
    
    G.add_nodes_from(nodes)
    
    for i in range(len(nodes)):
        G.nodes[nodes[i]]['category'] = node_labels[i]
    
    node_colors = []
    for i in range(len(nodes)):
        if node_labels[i] == 'object':
            node_colors.append('red')
        if node_labels[i] == 'positioning':
            node_colors.append('green')
        if node_labels[i] == 'action':
            node_colors.append('blue')
        if node_labels[i] == 'determinant':
            node_colors.append('orange')
    
    edges = []
    try:
        for i in range(len(sent)):
            edges.append((list(sent.keys())[i],list(sent.keys())[i+1]))
    except (RuntimeError, TypeError, NameError, IndexError):
        pass
        
    G.add_edges_from(edges)

    return G, node_colors
```

```python
graphs_TA = [graph(item) for item in list_of_wordsInSent_TA]
graphs_RA = [graph(item) for item in list_of_wordsInSent_RA]
```
